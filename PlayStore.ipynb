```python
#Libraries needed
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\sadaia academy\googleplaystore.csv")#read csv file
df.head()#show first 5 row by default
```

```python
#change the names because of the spaces between names. just add (_) between them 
df.rename(columns={'Last Updated':'Last_Updated'
                     , 'Content Rating':'Content_Rating'}, inplace = True)
df      
```

```python
df.shape # (number of rows, number of columns)
```

```python
df.describe()#summary statistic 
```

```python
df.dtypes #to know the type of the data to deal with
```

```python
duplicate = df[df.duplicated('App')]
duplicate
```

```python
df.info()
```

```python
# Rating column has NAN values so we have to handle it. I handle it using the mean of the column
df['Rating']=df['Rating'].fillna(df["Rating"].mean())
df.info()
```

```python
#We want to use 'Installs' as int so we have to remove the ('+') and (',')
df['Installs'] = df['Installs'].str.replace('+','')
df['Installs'] = df['Installs'].str.replace(',','')
df.head(3)
```

```python
df.info()
```

```python
#Convert 'Installs' to int 
df['Installs'] = df['Installs'].astype(int)
```

```python
df.info()
```

```python
df.Price.value_counts().unique()
```

```python
df.Price.value_counts()#price has ($) we have to clean it
```

```python
df['Price'] = df['Price'].str.replace('$','')
```

```python
df.Price.value_counts()
```

```python
df.info()
```

```python
df['Price'] = df['Price'].astype(float)
```

```python
df.info()
```

```python
#see if there are any duplicate data
duplicate =df.groupby(['App']).agg({'Category': 'count'}).reset_index()
duplicate = duplicate.sort_values(by=['Category'], ascending = False)#ascending = False. will sort it from the bigest to the smalest  
duplicate.head(50)
```

```python
roblox = df[df['App'].isin(['ROBLOX'])].sort_values(by=['Price'], ascending=True)#to make sure that all the data is the same
roblox
```

```python
before = len(df.index)
print('Total entries with duplicates were '+ (str(before)))
df.drop_duplicates(subset ="App",keep = 'first', inplace = True) #Drop duplicates with a value in a specific column.
after = len(df.index)
print('Total entries after duplicates were dropped are '+ (str(after)))
print('Total '+ (str(before-after))+' entries were dropped.')
```

```python
sns.boxplot(df['Rating'])
np.where(df['Rating']<3.3)
```

```python
#put a fixed number for the outliers (3.3)
df["Rating"] = np.where(df["Rating"] <3.3, 3.3,df['Rating'])
```

```python
# correlation between Rating and Installs "rocket_r" "vlag" "icefire" "Blues" "YlGnBu"
print(df['Rating'].corr(df['Installs']))
```

```python
forth=list(['Price','Installs','Rating'])
heatRelation = df[forth].corr()
ax = sns.heatmap(heatRelation, annot=True, cmap="icefire") #"rocket_r" "vlag" "icefire" "Blues" "YlGnBu" best color to represent
ax
```

```python
cat = df.groupby(['Category'])['App'].count().reset_index().rename(columns={'App':'Count of Apps'})
cat = cat.sort_values(by=['Count of Apps'], ascending=False)
cat.set_index('Category', inplace= True)
cat
```

```python
catInstalls = df.groupby(['Category'])['Installs'].sum().sort_values(ascending=False).reset_index()
catInstalls
```

```python
mergeCat = pd.merge(cat, catInstalls, on='Category')
mergeCat.sort_values(by=['Installs'], ascending=False)
```

```python
fig4, ax = plt.subplots(figsize =(20,3))
sns.barplot(x=catInstalls['Category'], y=catInstalls['Installs'])
plt.xlabel('\n Categories', fontsize = 10, fontweight = 'bold')
plt.ylabel('Installs', fontsize = 10, fontweight = 'bold')
plt.xticks(rotation = 90)
plt.title('Total Installs per Category', fontsize = 15, fontweight = 'bold')
plt.savefig('Total Installs per Category.png')
plt.show()
```

```python
Ageclass = df.groupby(['Content_Rating'])['Installs'].sum().sort_values(ascending=False).reset_index()
Ageclass
```

```python
#convert unrated to everyone 
df['Content_Rating'] = df['Content_Rating'].str.replace('Unrated','Everyone')
```

```python
Ageclass = df.groupby(['Content_Rating'])['Installs'].sum().sort_values(ascending=False).reset_index()
Ageclass
```

```python
fig4, ax = plt.subplots(figsize =(20,3))
sns.barplot(x=Ageclass['Content_Rating'], y=Ageclass['Installs'])
plt.xlabel('\n Content_Rating', fontsize = 10, fontweight = 'bold')
plt.ylabel('Installs by billions', fontsize = 10, fontweight = 'bold')
plt.title('Total Installs per age classification', fontsize = 15, fontweight = 'bold')
plt.savefig('Total Installs per Content_Rating.png')
plt.show()
```
